```python
import pandas as pd

df = pd.read_csv('sales_data_sample.csv', encoding='latin1')  # use 'latin1' to avoid encoding errors

print(df.shape)
df.head()
```

```python
df.info()

df.describe(include='all')
```

```python
print(df.isnull().sum())

df.dropna(inplace=True)  
```

```python
print("Duplicates before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)
```

```python
df['COUNTRY'] = df['COUNTRY'].str.strip().str.title()
df['STATUS'] = df['STATUS'].str.lower().str.strip()
df['PRODUCTLINE'] = df['PRODUCTLINE'].str.strip().str.title()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['quantityordered'] = df['quantityordered'].astype(int)
df['priceeach'] = df['priceeach'].astype(float)
df['sales'] = df['sales'].astype(float)
```

```python
df.to_csv('cleaned_sales_data.csv', index=False)
```
